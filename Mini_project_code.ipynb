```python
# Preprocessing the gathered Data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from skimage.io import imread
from skimage.transform import resize
 
data_folder ="/content/drive/MyDrive/Myinternship2021/DataScience/Mini project/Input"
categories =['Male','Female']
data=[]
target=[]
for i in categories:
  data_category=os.path.join(data_folder,i)
  #print(input_category)
  for  image in os.listdir(data_category):
    print(image)
    image_array=imread(os.path.join(data_category,image))
    image_rsize =resize(image_array,(1000,1000,3))
    data.append(np.ndarray.flatten(image_rsize))
    target.append(i)
    plt.imshow(image_rsize)
    plt.show()
data=np.array(data)
target=np.array(target)
```

```python
# Creating the DataFrame
df_data=pd.DataFrame(data)
df_data['OP'] = target
df_data
```

```python
# creating the x and y values
x=df_data.iloc[:,df_data.columns!='OP'].values
y=df_data.iloc[:,-1].values
```

```python
# Spilitting the input data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
```

```python
#Building the model using SVM algorithm 
from sklearn.svm import SVC
model = SVC()
model.fit(x_train,y_train)
y_p = model.predict(x_test)
y_p
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_p,y_test))
```

```python
#Creating the parameters for GridSearchCV
params= {'C':[1,10,100,1000],
         'gamma':[0.001,0.0001],
         'kernel':['rbf','linear','poly']}
```

```python
#Building the model using SVM algorithm with Hyperparamter tuning (using GridSearchCV)
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
grid = GridSearchCV(SVC(),params)
grid.fit(x_train,y_train)
```

```python
y_pred= grid.predict(x_test)
y_pred
```

```python
#Checking the accuracy,precision,recall and confusion matrix of the model
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print(f"Accuracy_Score:{accuracy_score(y_p,y_test)}")
print(f"Classification_Report:{classification_report(y_p,y_test)}")
print(f"Confusion_Matrix:{confusion_matrix(y_p,y_test)}")
```
